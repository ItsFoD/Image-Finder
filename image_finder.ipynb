```python
import requests
import os
import re
import random
import urllib
import urllib.request
import urllib.parse
import urllib.error
from bs4 import BeautifulSoup
```

```python
def image_finder():

    # user agent
    user_agent = 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'

    # get the search query
    search_query = input('Enter the search query: ')

    # get the amount of images to download
    amount_of_images = int(input('Enter the amount of images to download: '))

    # get the directory to save the images
    directory = input('Enter the directory to save the images: ')

    # get the url
    url = 'https://www.google.com/search?tbm=isch&q=' + search_query 

    # get the headers
    headers = {'User-Agent': user_agent}

    # get the request
    request = urllib.request.Request(url, None, headers)

    # get the response
    response = urllib.request.urlopen(request)

    # get the html
    html = response.read()

    # get the soup
    soup = BeautifulSoup(html, 'html.parser')

    # get the images
    images = soup.find_all('img')

    # get the image links
    image_links = [image['src'] for image in images]

    # get the image names
    image_names = [image['alt'] for image in images]

    # create the directory if it does not exist
    if not os.path.exists(directory):
        os.makedirs(directory)

    # download the images
    for i in range(amount_of_images):
        try:
            image = image_links[i]
            name = image_names[i]
            name = re.sub(r'[^A-Za-z0-9]+', '', name)
            name = name + str(random.randint(1, 100)) + '.jpg'
            image = requests.get(image)
            with open(directory + '/' + name, 'wb') as file:
                file.write(image.content)
                print('Downloaded', name)
        except:
            pass

def delete_directory():
    # Get the directory
    directory = input('Enter the directory to delete: ')

    if os.path.exists(directory):
        # Check the contents of the directory
        contents = os.listdir(directory)

        if len(contents) > 0:
            print('The directory is not empty:')
            for content in contents:
                print(content)

            # Prompt for confirmation after showing contents
            response = input('Do you want to delete the directory? (yes/no): ')
            if response == 'yes':
                # Remove the contents of the directory
                for content in contents:
                    os.remove(directory + '/' + content)
                # Remove the directory
                os.rmdir(directory)
                print('Directory deleted')
            else:
                print('Directory not deleted')
        else:
            os.rmdir(directory)
            print('Directory deleted')
    else:
        print('The directory does not exist')
```

```python
image_finder()
```

```python
delete_directory()
```
